```python
df.head(10)
```

```python
df['Education Rank'].head(10)
```

```python
df['Institution'].head()
```

```python
d=df.groupby(['Institution','Education Rank'])[['World Rank','National Rank','Employability Rank','Research Rank']].sum().reset_index()
```

```python
d.head(10)
```

```python
d.head(20)
```

```python
Institution =d['Institution'].unique()
```

```python
len(Institution)
```

```python
for idx in range(0,len(Institution)):
    D=d[d['Institution']==Institution[idx]].reset_index()
    plt.scatter(np.arange(0,len(D)),D['National Rank'],color='red',lable= 'National Rank')
    plt.title(Institution[idx])
    plt.xlable('Name Of Institutions')
    plt.ylable('Rank For Institutions')
    plt.legend()
    plt.show()
```
